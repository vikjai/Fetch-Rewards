### Splitting a column of dictionaries into separate columns with pandas.

```python
users["id"]=users['_id'].apply(pd.Series)
users.drop(['_id'], axis=1, inplace=True)
users.head()
```

```python
users["CreatedDate"]=users['createdDate'].apply(pd.Series)
users.drop(['createdDate'], axis=1, inplace=True)
users.head()
```

```python
users_lastLogin=users['lastLogin'].apply(pd.Series)
users_lastLogin
```

```python
users_lastLogin.columns
```

```python
users_lastLogin.drop(0, axis=1, inplace=True)
users['LastLogin'] = users_lastLogin
users.drop(['lastLogin'], axis=1, inplace=True)
users.head()
```

```python
receipts.head()
```

```python
receipts["id"]=receipts['_id'].apply(pd.Series)
receipts.drop(['_id'], axis=1, inplace=True)
receipts.head()
```

```python
receipts["CreateDate"]=receipts['createDate'].apply(pd.Series)
receipts.drop(['createDate'], axis=1, inplace=True)
receipts.head()
```

```python
receipts["DateScanned"]=receipts['dateScanned'].apply(pd.Series)
receipts.drop(['dateScanned'], axis=1, inplace=True)
receipts.head()
```

```python
receipts_finishedDate=receipts['finishedDate'].apply(pd.Series)
receipts_finishedDate.drop(0, axis=1, inplace=True)
receipts['FinishedDate'] = receipts_finishedDate
receipts.drop(['finishedDate'], axis=1, inplace=True)
receipts.head()
```

```python
receipts["ModifyDate"]=receipts['modifyDate'].apply(pd.Series)
receipts.drop(['modifyDate'], axis=1, inplace=True)
receipts.head()
```

```python
receipts_pointsAwardedDate=receipts['pointsAwardedDate'].apply(pd.Series)
receipts_pointsAwardedDate.drop(0, axis=1, inplace=True)
receipts['PointsAwardedDate'] = receipts_pointsAwardedDate
receipts.drop(['pointsAwardedDate'], axis=1, inplace=True)
receipts.head()
```

```python
receipts_purchaseDate=receipts['purchaseDate'].apply(pd.Series)
receipts_purchaseDate.drop(0, axis=1, inplace=True)
receipts['PurchaseDate'] = receipts_purchaseDate
receipts.drop(['purchaseDate'], axis=1, inplace=True)
receipts.head()
```

```python
brands.head()
```

```python
brands["id"]=brands['_id'].apply(pd.Series)
brands.drop(['_id'], axis=1, inplace=True)
brands.head()
```

```python
brands_cpg = brands['cpg'].apply(pd.Series)
brands = pd.concat([brands, brands_cpg], axis='columns')
brands.head()
```

```python
brands["oid"]=brands['$id'].apply(pd.Series)
brands.drop(['cpg', '$id'], axis=1, inplace=True)
brands.head()
```

### Converting UNIX epoch to datetime.

```python
users['CreatedDate'] = pd.to_datetime(users['CreatedDate'], unit='ms')
users['LastLogin'] = pd.to_datetime(users['LastLogin'], unit='ms')
users.head()
```

```python
receipts.columns
```
